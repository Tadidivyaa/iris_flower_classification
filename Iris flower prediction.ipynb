# Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=30)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

# Ridge Regression

```python
from sklearn.model_selection import GridSearchCV # GridSearchCV is a technique for finding the optimal parameter values from a given set of parameters in a grid
from sklearn.linear_model import Ridge
alpha = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20,30]

ridge = Ridge()
parameters = {'alpha': alpha}

ridge_regressor = GridSearchCV(ridge, parameters)
ridge_regressor.fit(X_train, y_train)
```

```python
ridge_regressor.best_params_#it finds the best parameter
```

```python
ridge=Ridge(alpha=1)
ridge.fit(X_train,y_train)
y_pred_ridge=ridge.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
Ridge_Error=mean_squared_error(y_pred_ridge,y_test)
Ridge_Error
```
